**Check if A and B are invertible by computing their invertible, if so, find inverse of A and the inverse of B**

To get the inverse of a matrix, 
- We need to find the determinant of the matrix

_Given matrix A_
    
    [[2, -1],
    [0, 3]]

det(A) = (2 * 3) - (-1 * 0)
       = 6 - 0 
       = 6

- _Since the det(A) != 0, A is invertible_

- _To find the inverse of A_:

    1/6[[3, 1], [-0, 2]]

    = 
    
      [[(1/6 * 3), (1/6 * 1)],
       [(1/6 * 0), (1/6 * 2)]]

 The inverse of A = 
    
      [[1/2, 1/6],
       [0, 1/3]]

_Given matrix B_
    
    [[1, 4],
     [-2, 5]]

det(B) = (1 * 5) - (4 * -2)
       = 5 - (-8) = 5 + 8
       = 13

- _Since the det(B) != 0, B is invertible_

- _To find the inverse of B_:

    1/13[[1, 4], [-2, 5]]

    = 
    
      [[(1/13 * 5), (1/13 * -4)],
       [(1/13 * 2), (1/13 * 1)]]

  The inverse of B = 
    
      [[5/13, -4/13],
