**Conditional Filtering**

```python
# Filter rows where Gender is 'Female'
filtered_male = df[df['Gender'] == 'Male'].head()
print("Rows where Gender is Male: ")
filtered_male
```

```python
# Filter rows where City is 'Lagos' and Course_Track is 'Data Science'

filtered_city = df[(df['City'] == 'Lagos') & (df['Course Track'] == 'AI')]
print("Rows where City is 'Lagos' and Course Track is 'Data Science':")
filtered_city
```

```python
# Filter rows where City is either 'Lagos' or 'Abuja'
cities = ['Lagos', 'Abuja']
city_filtered = df[df['City'].isin(cities)]
print("Rows where City is either 'Lagos' or 'Abuja':")
city_filtered
```

```python
# Using the query method
query_filtered = df.query("'Course Track' == 'AI' and Feedback == 'Excellent'")
print("Rows filtered using query() method where Course Track is AI and Feedback is Excellent: ")
query_filtered
```
